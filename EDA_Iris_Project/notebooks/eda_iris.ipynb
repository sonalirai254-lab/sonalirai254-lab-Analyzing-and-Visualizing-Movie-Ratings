# 🌸 Exploratory Data Analysis on the Iris Dataset

This notebook walks through a complete EDA pipeline:
1. Load & inspect
2. Clean
3. Describe
4. Correlate
5. Detect outliers
6. Visualize distributions & relationships
7. Summarize per species

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='Set2')
df = pd.read_csv('../data/iris.csv')
df.head()
```

## 1. Inspection

```python
print('Shape:', df.shape)
df.info()
print('Missing:\n', df.isnull().sum())
print('Duplicates:', df.duplicated().sum())
```

## 2. Cleaning

```python
df = df.drop_duplicates().reset_index(drop=True)
df.shape
```

## 3. Descriptive statistics

```python
df.describe().round(3)
```

```python
df['species'].value_counts()
```

## 4. Correlation heatmap

```python
num_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(6, 5))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

**Insight:** petal_length & petal_width are very strongly correlated (~0.96).

## 5. Outliers (IQR)

```python
for col in num_cols:
    q1, q3 = df[col].quantile([.25, .75])
    iqr = q3 - q1
    lo, hi = q1 - 1.5*iqr, q3 + 1.5*iqr
    n = ((df[col] < lo) | (df[col] > hi)).sum()
    print(f'{col}: {n} outliers')
```

## 6. Distributions

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 7))
for ax, col in zip(axes.ravel(), num_cols):
    sns.histplot(data=df, x=col, hue='species', kde=True, element='step', ax=ax)
plt.tight_layout(); plt.show()
```

## 7. Boxplots by species

```python
fig, axes = plt.subplots(1, 4, figsize=(14, 4))
for ax, col in zip(axes, num_cols):
    sns.boxplot(x='species', y=col, data=df, ax=ax)
plt.tight_layout(); plt.show()
```

## 8. Pairplot

```python
sns.pairplot(df, hue='species', diag_kind='kde')
plt.show()
```

## 9. Species-wise mean values

```python
df.groupby('species')[num_cols].mean().round(3)
```

## ✅ Conclusion
- Dataset is clean.
- Petal features dominate species discrimination.
- *Setosa* is linearly separable from the other two species.
- Logical next step: train a simple classifier (KNN / Logistic Regression).
